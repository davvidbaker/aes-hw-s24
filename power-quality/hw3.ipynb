```python
import pandas as pd
from functools import reduce
```

```python
df = pd.DataFrame(
    {
        "Harmonic #": [5, 7, 11, 13, 17, 19, 23, 25],
        "% of fundamental": [17.5, 11, 4.5, 2.9, 1.5, 1.0, 0.09, 0.08],
    }
)

df["magnitude [A]"] = df["% of fundamental"] * 173 / 100

display(df)
sum = reduce(lambda acc, x: acc + (x**2), df["magnitude [A]"], 0)
print(sum)
```
