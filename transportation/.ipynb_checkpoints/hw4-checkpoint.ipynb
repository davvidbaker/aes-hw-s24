```python
import matplotlib.pyplot as plt
import numpy as np
from hw_utils import nice_grid

%config InlineBackend.figure_format = 'retina'

fig, ax = plt.subplots(figsize=(10, 4))

fc_rail = 700  # [$]
fc_truck = 200  # [$]
energy_density_diesel = 38e6  # [J/L]
load = 20  # [tonnes]
diesel_price = 1.40  # [$/L]
energy_intensity_rail_base = 192.9e3  # [J/tkm]
energy_intensity_truck_base = 1788.3e3  # [J/tkm]

carbon_dioxide_emissions = 2.68  # [kg/L diesel]


# efficiency_improvements in percent
def energy_intensity_rail(efficiency_improvements):
    return energy_intensity_rail_base * ((100 - efficiency_improvements) / 100)


# efficiency_improvements in percent
def energy_intensity_truck(efficiency_improvements):
    return energy_intensity_truck_base * ((100 - efficiency_improvements) / 100)


# efficiency_improvements in percent
# carbon tax in $/tonne
def vc_rail(carbon_tax, efficiency_improvements):
    base_cost = (
        load
        * energy_intensity_rail(efficiency_improvements)
        * (1 / energy_density_diesel)
        * diesel_price
    )  # [$/km]
    carbon_cost = (
        load
        * energy_intensity_rail(efficiency_improvements)
        * (1 / energy_density_diesel)
        * carbon_tax
        * carbon_dioxide_emissions
        / 1000
    )  # [$/km]

    return base_cost + carbon_cost


# efficiency_improvements in percent
# carbon tax in $/tonne
def vc_truck(carbon_tax, efficiency_improvements):
    base_cost = (
        load
        * energy_intensity_truck(efficiency_improvements)
        * (1 / energy_density_diesel)
        * diesel_price
    )  # [$/km]

    carbon_cost = (
        load
        * energy_intensity_truck(efficiency_improvements)
        * (1 / energy_density_diesel)
        * carbon_tax
        * carbon_dioxide_emissions
        / 1000
    )  # [$/km]

    return base_cost + carbon_cost


# efficiency_improvements in percent
def breakeven(carbon_tax, efficiency_improvements):
    return (fc_rail - fc_truck) / (
        vc_truck(carbon_tax, efficiency_improvements)
        - vc_rail(carbon_tax, efficiency_improvements)
    )


carbon_tax = np.arange(0, 100.1, 0.1)

ax.plot(carbon_tax, breakeven(carbon_tax, 0), label="baseline")
ax.plot(carbon_tax, breakeven(carbon_tax, 10), label="with 10% efficiency improvements in engines")
ax.set_xlim(0, max(carbon_tax) + 0.1)
ax.set_ylim(0, 500)
ax.set_xlabel("carbon tax [$ \\frac{\$}{\mathrm{kg_{CO_2}}}$]")
ax.set_ylabel("breakeven distance [km]")
ax.legend()
nice_grid(ax)
```

We see that an increase in engine efficiency results in a larger breakeven distance. This makes sense because the improvement to engine efficiency reduces the variable costs of both transport modes, so the disparity in fixed cost has a larger effect.
